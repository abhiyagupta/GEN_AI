# Rephrase & Respond

image.png

In the Rephrase & Respond method of prompt engineering, we ask the LLM to rephrase the original user question to a format that will enable it to answer better.

The rephrase prompt looks like so:

```{prompt}
{question}
Given the above question, rephrase and expand it to help you do better answering.
Maintain all the information in the original question.
```

The final prompt looks like so:

```prompt
(original) {question}
(rephrased) {rephrased_question}
Use your answer for the rephrased question to answer the original question.
```

This approach is valuable for RAG-based question answering. Best performance is observed for Open AI's models; questions rephrased using GPT-4 can be used to prompt open source models.

# Chain-of-Verification (CoVe)

image.png

In CoVe, we prompt the LLM for a baseline response. Given this response, we generate a series of verification questions that are then presented to the LLM for consistency checks. Finally, we present the original question, baseline response, and the verification question-answer pairs to the LLM and prompt it to come up with a better-than-baseline answer.
